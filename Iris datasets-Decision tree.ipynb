```python
from sklearn.datasets import load_iris
import random
from math import log 
import pandas as pd
```

```python
data = load_iris()
# The data you read in is a dictionary. Look at the key and value of the dictionary to understand the data set
df1 = pd.DataFrame(data['data'], columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
df2 = pd.DataFrame(data['target'], columns=['Species'])
# Merge two dataframes with concat to get the target result
df = pd.concat([df1, df2], axis=1)
df
```

```python
def init_data_set():
    iris = load_iris()  
    iris_feature = iris.data.tolist()    #feature data
    iris_target = iris.target.tolist()   #classification data
    for i in range(len(iris_feature)):
        iris_feature[i].append(iris_target[i])
    return iris_feature
```

```python
# Partitioning the dataset
def create_set(data_set,split_rate=0.8):
    #0 is the test set，1 is the train set
    length=len(data_set)
    train_num=int(length*split_rate)
    test_num=length-train_num
    random_list=[1]*train_num
    random_list.extend([0]*test_num)
    random.shuffle(random_list)
    test_set=[]
    train_set=[]
    for i in range(length):
        if random_list[i]==0:
            test_set.append(data_set[i])
        else:
            train_set.append(data_set[i])
    print(test_set)
    print('----------divider---------------')
    print(train_set)
    return test_set,train_set
data_set=init_data_set()
create_set(data_set)
```

Calculate the Entropy:

```python
def calculate_Ent(data_set):
    #Create a dictionary for a category to count the number of categories
    label={}
    for i in data_set:
        if i[-1] not in label.keys():
            label[i[-1]]=1
        else:
            label[i[-1]]+=1
    #Calculate the Entropy
    Ent=0.0
    for i in label:
        tmp=float(label[i])/len(data_set)
        Ent-=tmp*log(tmp,2)
    print(Ent)
    return Ent
```

```python
#It is divided into two data sets based on pos and value
def spliit_Set(data_set,pos,value):
    less_Set=[]
    more_Set=[]
    for item in data_set:
        if item[pos]<value:
            less_Set.append(item)
        else:
            more_Set.append(item)
    return less_Set,more_Set
```

```python
#According to the criterion of max information gain, the optimal attribute is selected
def choose_best_split(data_set):
    #The information entropy of the initial case, that is, the information entropy of all data without partition
    base_Ent=calculate_Ent(data_set)
    #Record the maximum information gain and its corresponding attributes
    best_increase=0.0
    best_feature=[-1,-1]
    #Traverse each attribute, calculate the information gain, and find the maximum value
    for i in range(4):
        #Gets all the values of the property
        features=[j[i] for j in data_set]
        unique=set(features)
        #Gets all possible values of the property
        for feature in unique:
            less_Set,more_Set=spliit_Set(data_set, i, feature)
            tmp=len(less_Set)/float(len(data_set))
            new_Ent=tmp*calculate_Ent(less_Set)
            new_Ent+=(1-tmp)*calculate_Ent(more_Set)
            # Calculate infoGain
            increase=base_Ent-new_Ent
            # upgrade infoGain
            if increase>best_increase:
                best_increase=increase
                best_feature=[i,feature]
    #Returns the maximum information gain, the optimal attribute
    return best_feature,best_increase
```

```python
#Constrcute decision tree
def create_tree(data_set):
    myTree={}
    label=[i[-1] for i in data_set]
    label_set=set(label)
    #Sort by Value
    if len(label_set)==1:
        myTree['class']=label[0]
        # Returns the key with the most occurrences
        return myTree
    best_feature,best_increase=choose_best_split(data_set)
    myTree['node']=best_feature
    less_Set,more_Set=spliit_Set(data_set, best_feature[0], best_feature[1])
    myTree['left']=create_tree(less_Set)
    myTree['right']=create_tree(more_Set)
    return myTree
```

```python
#Draw the decision tree dictionary
def draw_tree(data_set):
    myTree={}
    label=[i[-1] for i in data_set]
    label_set=set(label)
    # When all LabelList are the same label
    if len(label_set)==1:
        return 'type:'+str(label[0])+'\nsample:'+str(len(data_set))
    # Obtain the best partition attribute and maximum information gain
    best_feature,best_increase=choose_best_split(data_set)
    string='X['+str(best_feature[0])+']<'+str(best_feature[1])
    string+='\nbest_increase='+str(round(best_increase,3))
    string+='\nsample:'+str(len(data_set))
    # Constrcute tree
    myTree[string]={}
    # The best partition attribute - BestFeature is removed from the attribute set
    less_Set,more_Set=spliit_Set(data_set, best_feature[0], best_feature[1])
    # The recursive call creates the decision tree
    myTree[string]['True']=draw_tree(less_Set)
    myTree[string]['False']=draw_tree(more_Set)
    return myTree 
```

```python
#Once have the dictionary of the decision tree,use python to draw the corresponding decision tree, figure
import matplotlib.pyplot as plt
#Define the shape of the judged node, where boxstyle is the text box type and fc is the depth of the comment box color
decisionNode = dict(boxstyle="round4", color='r', fc='0.9')
#Define the leaf shape
leafNode = dict(boxstyle="circle", color='m')
#Defines the shape of an arrow from a parent node to a child node or leaf
arrow_args = dict(arrowstyle="<-", color='g')
def plot_node(node_txt, center_point, parent_point, node_style):
  
    createPlot.ax1.annotate(node_txt, 
                            xy=parent_point,
                            xycoords='axes fraction',
                            xytext=center_point,
                            textcoords='axes fraction',
                            va="center",
                            ha="center",
                            bbox=node_style,
                            arrowprops=arrow_args)
```

```python
def get_leafs_num(tree_dict):
    
    # The total number of leaf nodes in the tree_dict
    leafs_num = 0
    
    # The first key in the dictionary, which is the first node in the tree
    root = list(tree_dict.keys())[0]
    # The values corresponding to this key are all subtrees of that node.
    child_tree_dict =tree_dict[root]
    for key in child_tree_dict.keys():
        if type(child_tree_dict[key]).__name__=='dict':
            #If a subtree is dic-typical, the number of leaves in the current tree is added to the number of leaves in this subtree
            leafs_num += get_leafs_num(child_tree_dict[key])
        else:
            leafs_num += 1
    return leafs_num
```

```python
def get_tree_max_depth(tree_dict):
    max_depth = 0
    root = list(tree_dict.keys())[0]
    child_tree_dict = tree_dict[root]
    for key in child_tree_dict.keys():
        # The number of levels of the current branch of the tree
        this_path_depth = 0
        if type(child_tree_dict[key]).__name__ == 'dict':
            #If the subtree is word typical, the number of levels of the current branch needs to be added to the deepest level of the subtree
            this_path_depth = 1 + get_tree_max_depth(child_tree_dict[key])
        else:
            this_path_depth = 1
        if this_path_depth > max_depth:
            max_depth = this_path_depth
    return max_depth
```

```python
def plot_mid_text(center_point, parent_point, txt_str):
    x_mid = (parent_point[0] - center_point[0])/2.0 + center_point[0]
    y_mid = (parent_point[1] - center_point[1])/2.0 + center_point[1]
    createPlot.ax1.text(x_mid, y_mid, txt_str)
    return
```

```python
def plotTree(tree_dict, parent_point, node_txt):

    leafs_num = get_leafs_num(tree_dict)
    root = list(tree_dict.keys())[0]
    center_point = (plotTree.xOff+(1.0+float(leafs_num))/2.0/plotTree.totalW,plotTree.yOff)
    #fill node_txt context
    plot_mid_text(center_point, parent_point, node_txt)
    plot_node(root, center_point, parent_point, decisionNode)
    child_tree_dict = tree_dict[root]
    plotTree.yOff=plotTree.yOff-1.0/plotTree.totalD
    for key in child_tree_dict.keys():
        if type(child_tree_dict[key]).__name__ == 'dict':
            plotTree(child_tree_dict[key],center_point,str(key))
        else:
            plotTree.xOff=plotTree.xOff+1.0/plotTree.totalW
            plot_node(child_tree_dict[key],(plotTree.xOff,plotTree.yOff),center_point,leafNode)
            plot_mid_text((plotTree.xOff,plotTree.yOff),center_point,str(key))
    plotTree.yOff=plotTree.yOff+1.0/plotTree.totalD
    return
```

```python
def createPlot(tree_dict):
    plt.figure(figsize=(20,8)) 
    fig=plt.figure(1,facecolor='white')
    fig.clf()
    axprops = dict(xticks=[], yticks=[])
    createPlot.ax1=plt.subplot(111, frameon=False, **axprops)
    plotTree.totalW=float(get_leafs_num(tree_dict))
    plotTree.totalD=float(get_tree_max_depth(tree_dict))
    plotTree.xOff=-0.5/plotTree.totalW;
    plotTree.yOff=1.0;
    plotTree(tree_dict, (0.5,1.0), '')
    plt.show()
```
